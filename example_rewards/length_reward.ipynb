```python
import numpy as np
import matplotlib.pyplot as plt

from .length_reward import asymmetric_linear_with_plateau

TARGET_LEN = 2000
PLATEAU = 200
SIGMA = 500
STEEPNESS = 0.002

lengths = np.linspace(0, 5000, 1000)
rewards = [asymmetric_linear_with_plateau(length, TARGET_LEN, PLATEAU, 0.5) for length in lengths]

plt.figure(figsize=(12, 6))
plt.plot(lengths, rewards)
plt.title('Completion Length vs Reward')
plt.xlabel('Completion Length (characters)')
plt.ylabel('Reward')
plt.grid(True)

plt.axvline(x=TARGET_LEN, color='g', linestyle='--', label=f'Target Length ({TARGET_LEN})')

plt.legend()
plt.show()
```
